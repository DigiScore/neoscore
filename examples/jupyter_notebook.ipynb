# Embed Neoscore in Jupyter Notebooks

Neoscore has experimental support for rendering score images to Jupyter Notebooks. This notebook is a basic demonstration of how to do that.

First, we begin by running all our imports and initial setup code in a standalone cell.

```python
from neoscore.common import *
neoscore.setup()
```

Next, let's create an object and display it.

```python
test_text = Text(ORIGIN, None, "Hello, world!",
                 neoscore.default_font.modified(size=Unit(8)))
neoscore.render_to_notebook()
```

Since we kept a reference to the object we created, we can modify it and see what the score looks like after.

```python
test_text.brush = "#f00"
neoscore.render_to_notebook()
```

Just like in the interactive REPL and animated environment, we can remove objects by explicitly calling `remove`.

```python
test_text.remove()
```

Notice how when we don't call `render_to_notebook`, no output is displayed.

Now let's try drawing a staff.

```python
staff = Staff(ORIGIN, None, Mm(100))
clef = Clef(ZERO, staff, 'treble')
key_signature = KeySignature(ZERO, staff, 'ef_major')
neoscore.render_to_notebook()
```

And let's put some notes on it.

```python
Chordrest(staff.unit(2), staff, ["c", "f#", "e'"], (3, 16))
neoscore.render_to_notebook()
```

We can change the size of rendered images by tweaking the DPI parameter. You can also configure things like what area is exported and the image quality.

```python
neoscore.render_to_notebook(dpi=120)
```

[See the docs](https://neoscore.org/core_concepts/interactivity) for more on how to use this integration.
